```python
!pip install matplotlib
!pip install pandas
!pip install matplotlib-ganttimport matplotlib.pyplot as plt
import pandas as pd

# Sample data for Gantt chart
data = {
    "Task": ["Task 1", "Task 2", "Task 3"],
    "Start Date": ["2024-01-01", "2024-02-01", "2024-03-01"],
    "End Date": ["2024-01-15", "2024-02-15", "2024-03-15"],
}

df = pd.DataFrame(data)
df["Start Date"] = pd.to_datetime(df["Start Date"])
df["End Date"] = pd.to_datetime(df["End Date"])

# Plotting the Gantt chart
fig, ax = plt.subplots(figsize=(10, 5))

for i, row in df.iterrows():
    ax.barh(row["Task"], width=row["End Date"] - row["Start Date"], left=row["Start Date"])

ax.set_xlabel("Timeline")
ax.set_ylabel("Tasks")
ax.set_title("Sample Gantt Chart")
plt.show()
```
