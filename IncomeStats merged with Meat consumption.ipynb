```python
import pandas as pd

# Load the first CSV file
meat_consumption_df = pd.read_csv('modified_meat_consumption_worldwide.csv')

# Load the second CSV file
historical_data_df = pd.read_csv('Final_Historical_data_ISO.csv')

# Merge the two dataframes on the 'Country Name' and 'country' columns
merged_df = pd.merge(meat_consumption_df, historical_data_df, left_on='Country Name', right_on='country', how='inner')

# Save the merged dataframe to a new CSV file
merged_df.to_csv('merged_meat_consumption_and_Income stats.csv', index=False)

# Display the first few rows of the merged dataframe
merged_df.head()
```

```python
# Load the merged CSV file
merged_df = pd.read_csv('merged_meat_consumption_and_Income stats.csv')

# List of columns to drop
columns_to_drop = ['LOCATION', 'country', 'iso', 'REF', 'Data source']

# Drop the specified columns
merged_df.drop(columns=columns_to_drop, inplace=True)

# Save the updated dataframe to a new CSV file
merged_df.to_csv('merged_meat_consumption_and_Income_stats_withDroppedCoulmns.csv', index=False)

# Display the first few rows of the updated dataframe to confirm changes
merged_df.head()
```

```python
import pandas as pd

# Load the updated CSV file
df = pd.read_csv('merged_meat_consumption_and_Income_stats_withDroppedCoulmns.csv')

# Replace '#N/A' with NaN
df.replace('#N/A', pd.NA, inplace=True)

# Drop rows with NaN values
df.dropna(inplace=True)

# Save the cleaned dataframe to a new CSV file
df.to_csv('cleaned_merged_meat_consumption_and_Income_stats.csv', index=False)

# Display the first few rows of the cleaned dataframe to confirm changes
df.head()
```

```python
# Load the merged CSV file
merged_df = pd.read_csv('cleaned_merged_meat_consumption_and_Income_stats.csv')

# List of columns to drop
columns_to_drop = ['gini_recalculated', 'GCAM_region_ID']

# Drop the specified columns
merged_df.drop(columns=columns_to_drop, inplace=True)

# Save the updated dataframe to a new CSV file
merged_df.to_csv('UpdatedwithDroppedCoulmns_meat_consumption_and_Income_stats_withDroppedCoulmns.csv', index=False)

# Display the first few rows of the updated dataframe to confirm changes
merged_df.head()
```
